```python
from keras.models import load_model
model = load_model('79_accuracy_not_augmented.h5')
model.author = 'Anton Poluianov'
input_shape = {
    "image":[None, 224, 224, 3]
}
coreml_model = coremltools.converters.keras.convert(model,
                                                    input_names='image',
                                                    image_input_names='image',
                                                    input_name_shape_dict=input_shape,
                                                    output_names='probabilities',
                                                    class_labels=['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash'])

coreml_model.save('TrashClassifier79accNotAugmented.mlmodel')
print('The model was saved')
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
from keras.applications import imagenet_utils
from keras.preprocessing import image
import numpy as np

def load_image(img_path, show=False):

    img = image.load_img(img_path, target_size=(150, 150))
    img_tensor = image.img_to_array(img)                    # (height, width, channels)
    img_tensor = np.expand_dims(img_tensor, axis=0)         # (1, height, width, channels), add a dimension because the model expects this shape: (batch_size, height, width, channels)
    img_tensor /= 255.                                      # imshow expects values in the range [0, 1]

    if show:
        plt.imshow(img_tensor[0])                           
        plt.axis('off')
        plt.show()

    return img_tensor

def make_pred(img_path):
    new_image = load_image(img_path)
    pred = model.predict(new_image)
    print(np.argmax(pred))
    return pred

print('cardboard')
print(make_pred('./merged_full_split_final/train/cardboard/cardboard1.jpeg'))

print('glass')
print(make_pred('./merged_full_split_final/train/glass/glass1.jpeg'))

print('metal')
print(make_pred('./merged_full_split_final/train/metal/metal10.jpeg'))

print('paper')
print(make_pred('./merged_full_split_final/train/paper/paper10.jpeg'))

print('plastic')
print(make_pred('./merged_full_split_final/train/plastic/plastic144.jpeg'))

print('trash')
print(make_pred('./merged_full_split_final/train/trash/trash1.jpeg'))

```
